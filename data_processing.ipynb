### Label Encoding for Categorical Data

```python
dt.dtypes
```

```python
#change datatypes for string cols
cols=['Gender','Area','Loan_Status']
dt[cols]=dt[cols].astype('category')
```

```python
dt.dtypes
```

```python
for c in cols:
    dt[c]=dt[c].cat.codes
```

```python
dt
```
